# SARIMAX model for real GDP forecasting

Importing libraries

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm

from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt

import pandas as pd

import numpy as np

from tqdm import tqdm_notebook

from itertools import product

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [10, 7.5]
```

We will be looking at the United States Macroeconomic Dataset

```python
macro_data = sm.datasets.macrodata.load_pandas()
macro_data.data
```

```python
macro_data = macro_data.data
```

Let us see how the data looks like as a function of time

```python
#Visualize data
fig, axes = plt.subplots(nrows = 6, ncols = 2, dpi = 240, figsize = (10, 10))

for i, ax in enumerate(axes.flatten()):
    data = macro_data[macro_data.columns[i+2]]

    ax.plot(data, color = "black", linewidth = 1)
    ax.set_title(macro_data.columns[i+2])
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.spines['top'].set_alpha(0)
    ax.tick_params(labelsize = 6)

plt.tight_layout()



#Not interested in year and quarter columns, so plotting the 12 remaining
```

We can see that the time-series is not stationary. In an attempt to make it stationary, we will do differencing. 

```python
#Apply differencing to make time series stationary

macro_data['realgdp_tr1'] = macro_data['realgdp'].diff()

macro_data.head(10)
```

```python
plt.plot(macro_data['realgdp_tr1'])
plt.title('Transformed RealGDP')
plt.show()
```

We can see there is no trend, but the variance is quite high. So as to make sure that the variance is static for a stationary time series, we will perform the adfuller statisitical test.

```python
#No trend but high variance, needs to check stationarity with adfuller test (null hypothesis test)
ad_fuller_result = adfuller(macro_data['realgdp_tr1'][1:])

print(f'ADF Statistic: {ad_fuller_result[0]}')
print(f'p-value: {ad_fuller_result[1]}')
```

Since the p-value is less than 0.05, the time series is stationary

Let's now plot the ACF and PACF on this now stationary time series

```python
plot_acf(macro_data['realgdp_tr1'][1:])
plot_pacf(macro_data['realgdp_tr1'][1:])
```

Both the ACF and PACF are decaying and sinosoidal. Hence, they cannot provide any useful information. Let us now take the SARIMAX model and try to fit and see how it performs. 

We will be writing an optimization function to find the optimum combination of p,q (auto regression, time-differencing), and P, Q (seasonality parameters). We will be passing all the remaining variables as the exogenous variables. We assume the seasonality difference to be zero and differencing for the ARIMA part to be 1

```python
#ACF and PACF are both inconclusive, hence, need to optimize using SARIMAX optimization model
def optimize_SARIMAX(endog, exog, parameters_list, d, D, s):
    """
    Returns dataframe with parameters, corresponding AIC, 

    endog - the observed variable
    exog - the exogenous variables (or X)

    paramters_list - list with(p,d,P,Q)

    d - integration order

    D - Seasonal integration order

    s - length of the season
    """

    results = []

    for param in tqdm_notebook(parameters_list):
        try:
            model = SARIMAX(endog, 
            exog,
            order = (param[0], d, param[1]),
            seasonal_order=(param[2], D, param[3], s),
            simple_differencing= False).fit(disp = False)

        except:
            continue

        aic = model.aic
        results.append([param, aic])

    result_df = pd.DataFrame(results)

    result_df.columns = ['(p,q)x(P,Q)', 'AIC']

    result_df = result_df.sort_values(by = 'AIC', ascending = True).reset_index(drop = True)

    return result_df

```

```python
p = range(0, 4)

d = 1

q = range(0, 4)

P = range(0, 4)

D = 0

Q = range(0, 4)

s = 4  #Quarterly 

parameters = product(p,q,P,Q)
parameters_list =list(parameters)

```

```python
endog = macro_data['realgdp'][:200]  #Training Set, training on 1st 200 data points
exog = macro_data[['realcons', 'realinv', 'realgovt', 'realdpi', 'cpi', 'm1', 'tbilrate', 'unemp', 'pop', 'infl', 'realint']][:200]

result_df = optimize_SARIMAX(endog, exog, parameters_list, 1, 0, 4)

result_df
```

```python
result_df.head()
```

Based on AIC, the optimum parameters are p = 2, q = 2, P = 1, and Q = 1

We will now take these optimum parameters and fit the data with the SARIMAX model

```python
#Optimum values are p = 2, q = 2, P = 1, Q = 1
#Let's build a model from these optimum parameters

best_model = SARIMAX(endog,
                    exog,
                    order = (2,1,2),
                    seasonal_order=(1,0,0,4),
                    simple_differencing=False)

res = best_model.fit(dis = False)

print(res.summary())
```

Let us now look at the residual diagnostics to gauge how good the fit is to the actual data.

```python
#Study residuals, should follow a normal distribution

res.plot_diagnostics()
```

It can be seen that the residuals are very close to a normal distribution, showing that our model fit the data quite well.

Let's try to forecast the next 3 quarters 

```python
#Testing on unseen data
n_forecast = 3

predict = res.get_prediction(end = best_model.nobs + n_forecast, 
                                            exog = exog.iloc[-4:])

idx = np.arange(len(predict.predicted_mean))

#Confidence intervals

predict_ci = predict.conf_int(alpha = 0.5)

fig, ax = plt.subplots()

ax.plot(macro_data['realgdp'][:199], color = 'blue')
ax.plot(idx[-n_forecast:], predict.predicted_mean[-n_forecast:], 'k--')
ax.fill_between(idx, predict_ci.iloc[:,0], predict_ci.iloc[:,1], alpha = 0.15)

ax.set_title('Forecast of US Real GDP')
plt.show()
```

Let's zoom in on the last 9 observations and look at the confidence intervals to see how well we have fitted the curve.



```python
fig, ax = plt.subplots()

ax.plot(macro_data['realgdp'], color = 'blue')
ax.plot(idx[-n_forecast :], predict.predicted_mean[-n_forecast :], 'k--')
ax.plot(predict.predicted_mean, 'ro')

ax.fill_between(idx, predict_ci.iloc[:, 0], predict_ci.iloc[:,1], alpha = 0.15)

ax.set_xlim(190, 199)

ax.set_ylim(13000, 14100)

ax.set_title('Forecast of Real GDP')

plt.show()
```

It can be seen that our model fits the data quite nicely.

For a more quantified outlook as to how well our model is performing, let's look at MSE and Mean Percentage Error

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(macro_data['realgdp'], predict.predicted_mean[: 203])

print(f'MSE: {mse}')
```

```python
#MSE is large, let's calculate Mean percentage error

mape = np.mean(np.abs(macro_data['realgdp'] - predict.predicted_mean[:203]) / np.abs(macro_data['realgdp']))

print(mape)
```

With low mean percentage error of 0.38 %, it can be seen that our model predicts the data quite nicely.
